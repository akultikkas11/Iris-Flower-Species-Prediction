```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the csv file
iris_df = pd.read_csv('Iris.csv')
iris_df
```

```python
#Getting sample of data ->
iris_df.head()
```

```python
#Check whether the data is clean or not -->
iris_df.isnull().sum()
```

```python
#Now split the data into input and output variable --->

#Input variable -->
X = iris_df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
X
```

```python
#Output varaible -->
y = iris_df["Species"]
y
```

```python
#AS the output is a stirng value, we need to convert it into numerical value
#For this conversion we make the use of Label Encoding ---->

# 0 ---> setosa
# 1 ---> versicolor
# 2 ---> virginica

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y1 = le.fit_transform(y)
y1
```

```python
#new iris dataset --->
new_iris_df = iris_df.copy()
new_iris_df["Species Value"] = y1
new_iris_df.head()
```

```python
#Now we need to visualize the data, i.e. we need to visualize each species based on sepal and petal length and width
#So we group the species -->
species = new_iris_df.groupby("Species Value")
setosa = species.get_group(0)
versicolor = species.get_group(1)
virginica = species.get_group(2)
```

```python
#Now we visualize the data --->
import matplotlib.pyplot as plt
```

```python
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.scatter(x=setosa["SepalLengthCm"], y=setosa["SepalWidthCm"], color="r", label="setosa")
plt.scatter(x=versicolor["SepalLengthCm"], y=versicolor["SepalWidthCm"], color="b", label="versicolor")
plt.scatter(x=virginica["SepalLengthCm"], y=virginica["SepalWidthCm"], color="g", label="virginica")
plt.legend()
```

```python
#Petal Length vs Petal Width
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.scatter(x=setosa["PetalLengthCm"], y=setosa["PetalWidthCm"], label="setosa", color="r")
plt.scatter(x=versicolor["PetalLengthCm"], y=versicolor["PetalWidthCm"], label="versicolor", color="b")
plt.scatter(x=virginica["PetalLengthCm"], y=virginica["PetalWidthCm"], label="virginica", color="g")
plt.legend(loc="lower right")
```

```python
#Now creating the model --->
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
model = KNeighborsClassifier()
```

```python
#Splitting the data into training and testing dataset --->
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.2)

#Training the model --->
model.fit(X_train, y_train)

#Checking the score/accuracy of the model ---->
score = model.score(X_test, y_test)
print(f"Accuracy/score is : {score}")
```

```python
#app.py file for streamlit....
%%writefile app.py
import streamlit as st
import pickle
model1 = pickle.load(open("Iris.pkl", "rb"))

def my_model() :
    sepal_len = st.number_input("Enter sepal length : ")
    sepal_wid = st.number_input("Enter sepal width : ")
    petal_len = st.number_input("Enter petal length : ")
    petal_wid = st.number_input("Enter petal width : ")
    pred = st.button("Predict")
    result = None

    if(pred) :
        result = model1.predict([[sepal_len, sepal_wid, petal_len, petal_wid]])

    st.write("Flower species : ", result)

my_model()
```
